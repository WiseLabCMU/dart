```python
from dart import VirtualCameraImage
from matplotlib import pyplot as plt
import numpy as np
from scipy.io import loadmat
import h5py
```

```python
cam_real = VirtualCameraImage.from_file(
    "results/cabinets.iid/cam_all.mat").to_rgb()
cam_sim = VirtualCameraImage.from_file(
    "results/cabinets.sim.iid/cam_all.mat").to_rgb()
rad_real = loadmat("results/cabinets.iid/pred_all.mat")["rad"]
rad_sim = loadmat("results/cabinets.sim.iid/pred_all.mat")["rad"]
```

```python
fig, axs = plt.subplots(4, 8, figsize=(24, 12))
# idxs = np.random.choice(len(cam_real), 8)
idxs = [29, 89, 173, 371, 805, 1412, 1559, 1800]

for i, ax in zip(idxs, axs[0]):
    ax.set_title(i)
    ax.imshow(cam_real[i])
for i, ax in zip(idxs, axs[1]):
    ax.imshow(cam_sim[i])
for i, ax in zip(idxs, axs[2]):
    ax.imshow(rad_real[i], origin='lower')
for i, ax in zip(idxs, axs[3]):
    ax.imshow(rad_sim[i], origin='lower')

for ax in axs.reshape(-1):
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
```
